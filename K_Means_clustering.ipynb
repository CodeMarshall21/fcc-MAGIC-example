```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/income.csv")
df
```

```python
plt.scatter(df["Age"], df["Income($)"])
```

```python
km_model = KMeans(n_clusters = 3)
km_model
```

```python
y_predict = km_model.fit_predict(df[["Age"]], df[["Income($)"]])
y_predict
```

```python
df["cluster"] = y_predict
df
```

```python
df_cluster_0 = df[df["cluster"] == 0]
df_cluster_1 = df[df["cluster"] == 1]
df_cluster_2= df[df["cluster"] == 2]

plt.scatter(df_cluster_0["Age"], df_cluster_0["Income($)"], color = "green", label = "Cluster 0")
plt.scatter(df_cluster_1["Age"], df_cluster_1["Income($)"], color = "red", label = "Cluster 1")
plt.scatter(df_cluster_2["Age"], df_cluster_2["Income($)"], color = "black", label = "Cluster 2")

plt.xlabel("Age")
plt.ylabel("Income ($)")
plt.legend()
```

# Scaling

```python
min_max = MinMaxScaler()
df["Income($)"] = min_max.fit_transform(df[["Income($)"]])
df["Age"] = min_max.fit_transform(df[["Age"]])
df
```

```python
km_model = KMeans(n_clusters = 3)
y_pred = km_model.fit_predict(df[["Age","Income($)"]])

df["cluster"] = y_pred
df
```

```python
df_cluster_0 = df[df["cluster"] == 0]
df_cluster_1 = df[df["cluster"] == 1]
df_cluster_2= df[df["cluster"] == 2]

plt.scatter(df_cluster_0["Age"], df_cluster_0["Income($)"], color = "green", label = "Cluster 0")
plt.scatter(df_cluster_1["Age"], df_cluster_1["Income($)"], color = "red", label = "Cluster 1")
plt.scatter(df_cluster_2["Age"], df_cluster_2["Income($)"], color = "black", label = "Cluster 2")

plt.scatter(km_model.cluster_centers_[:,0], km_model.cluster_centers_[:,1], color = "purple", marker = "*", label = "centroid")

plt.xlabel("Age")
plt.ylabel("Income ($)")
plt.legend()
```

# Elbow Method

```python
k_range = range(1,10)
SSE = []

for k_clusters in k_range:
  km_model = KMeans(n_clusters = k_clusters)
  y_pred = km_model.fit_predict(df[["Age", "Income($)"]])
  SSE.append(km_model.inertia_)

SSE
```

```python
plt.plot(k_range, SSE)
plt.xlabel("K_clusters")
plt.ylabel("SSE")
```
