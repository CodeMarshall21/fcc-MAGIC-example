```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/salaries.csv")
df
```

```python
feature_matrix = df.drop("salary_more_then_100k", axis = 1)
target_vector = df["salary_more_then_100k"]
```

```python
feature_matrix
```

```python
target_vector
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
company_encoder = LabelEncoder()
job_encoder = LabelEncoder()
degree_encoder = LabelEncoder()
```

```python
feature_matrix["company_encoded"] = company_encoder.fit_transform(feature_matrix["company"])
feature_matrix["job_encoded"] = job_encoder.fit_transform(feature_matrix["job"])
feature_matrix["degree_encoded"] = degree_encoder.fit_transform(feature_matrix["degree"])

feature_matrix
```

```python
# feature_matrix = feature_matrix.drop(["company", "job",	"degree"], axis = "columns")
feature_matrix = feature_matrix.drop(feature_matrix.columns[:3], axis = "columns")
feature_matrix
```

```python
from sklearn.model_selection import train_test_split
```

```python
feature_train, feature_test, target_train, target_test = train_test_split(feature_matrix, target_vector, test_size = 0.25, random_state = 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
decisionTreeModel = DecisionTreeClassifier()
```

```python
decisionTreeModel.fit(feature_train, target_train)
```

```python
decisionTreeModel.score(feature_test, target_test)
```
